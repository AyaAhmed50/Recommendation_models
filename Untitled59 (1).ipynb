```python
# TF-IDF Vectorization on Product Names
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Product Name'])

# Compute Cosine Similarity Matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def get_content_based_recommendations(product_name, top_n=5):
    try:
        idx = df[df['Product Name'] == product_name].index[0]
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
        product_indices = [i[0] for i in sim_scores]
        return df.iloc[product_indices][['Product ID', 'Product Name']]
    except IndexError:
        return pd.DataFrame(columns=['Product ID', 'Product Name'])

# Example Usage
product_example = df['Product Name'].iloc[0]  # Select a sample product
recommended_products = get_content_based_recommendations(product_example)
print(f"Content-Based Recommendations for '{product_example}':")
print(recommended_products)

# Calculate Accuracy for Content-Based Model
actual_ratings = []
predicted_ratings = []

for index, row in df.iterrows():
    product_name = row['Product Name']
    actual_rating = row['Rate']
    recommended_products = get_content_based_recommendations(product_name, top_n=1)
    if not recommended_products.empty:
        predicted_rating = df[df['Product Name'] == recommended_products.iloc[0]['Product Name']]['Rate'].mean()
        actual_ratings.append(actual_rating)
        predicted_ratings.append(predicted_rating)

mse = mean_squared_error(actual_ratings, predicted_ratings)
rmse = np.sqrt(mse)
mae = mean_absolute_error(actual_ratings, predicted_ratings)
```

```python
# Combine product name with sub-category for better matching
df['Full Description'] = df['Product Name'] + ' ' + df['Sub-Category']

# Get unique full descriptions
product_texts = df[['Product ID', 'Full Description']].drop_duplicates().reset_index(drop=True)

# TF-IDF vectorization
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(product_texts['Full Description'])

# Sample customers for evaluation (use more for better testing)
sample_customers = df['Customer ID'].unique()[:10]  # Evaluate on 10 customers
total_precision = []

# Simulate a search and evaluate recommendations for each customer
for customer_id in sample_customers:
    # Simulate a search by picking a random product the customer bought
    customer_products = df[df['Customer ID'] == customer_id]
    if customer_products.empty:
        continue

    # Pick one product as the "search query"
    searched_product = customer_products.sample(1)['Full Description'].values[0]

    # Vectorize the search query
    search_vector = vectorizer.transform([searched_product])

    # Compute cosine similarity between search and all products
    cosine_similarities = cosine_similarity(search_vector, tfidf_matrix).flatten()

    # Get top 10 most similar products (excluding the original product itself)
    top_indices = cosine_similarities.argsort()[-11:][::-1]  # top 11 to exclude the same product
    recommended_ids = [product_texts.iloc[i]['Product ID'] for i in top_indices if product_texts.iloc[i]['Full Description'] != searched_product][:10]

    # Get actual products purchased by the customer (excluding the searched one)
    actual_purchased_ids = customer_products['Product ID'].unique()

    # Calculate precision
    true_positives = len(set(recommended_ids) & set(actual_purchased_ids))
    precision = true_positives / len(recommended_ids) if recommended_ids else 0
    total_precision.append(precision)

    print(f"Customer: {customer_id} | Precision: {precision:.2f}")

# Compute average precision across customers
avg_precision = np.mean(total_precision)
print("\nAverage Precision across all tested customers:", round(avg_precision, 2))
```
